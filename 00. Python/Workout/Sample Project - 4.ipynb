```python
import json

class FileHandling:
    def __init__(self, inFile, outFile):
        self.inputFile = inFile
        self.outputFile = outFile
    
    def readJSONFile(self):
        with open(self.inputFile) as fileObj:
            fileContent = json.load(fileObj)
        return fileContent
        
    def writeJSONFile(self, dataDict):
        with open(self.outputFile, 'w') as fileObj:
            json.dump(dataDict, fileObj, indent=4)
        
```

```python
class ResultMangement(FileHandling):
    passMark = 45
    def __init__(self, inFile, outFile):
        FileHandling.__init__(self, inFile, outFile)
        self.stdData = self.readJSONFile() 
    
    @staticmethod
    def findTotal(marks):
        try:
            return sum(marks)
        except Exception as e:
            return 0
        
    @classmethod 
    def findResult(cls, marks):
        try:
            for mark in marks:
                if mark < cls.passMark:
                    return 'Fail'
                    break
            else:
                return 'Pass'
        except Exception as e:
            return '' 
        
    @staticmethod
    def findAvg(marks):
        try:
            return sum(marks)/len(marks)
        except Exception as e:
            return 0
    
    def calResult(self):
        for std in self.stdData:
            std['Total'] = ResultMangement.findTotal(std['marks'])
            std['Result'] = ResultMangement.findResult(std['marks'])
            if std['Result'] == 'Pass':
                std['Avg'] = ResultMangement.findAvg(std['marks'])
        
        self.writeJSONFile(self.stdData) 
    
```

```python
stdObj = ResultMangement('../Files/Students-raw.json', '../Files/output/students-class.json')
stdObj.calResult()
```

```python
class IPLMangement(FileHandling):
    def __init__(self, inFile, outFile):
        FileHandling.__init__(self, inFile, outFile)
        self.stdData = self.readJSONFile() 
    
```
