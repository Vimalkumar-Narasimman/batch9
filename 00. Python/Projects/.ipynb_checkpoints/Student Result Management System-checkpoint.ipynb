### Student Result Management System using JSON and Class concepts

1. Read <code>student.json</code> to calculate below:
    1. Find a total mark for each student
    2. Find result of the student, pass mark is 45
    3. If student passed all subject then calculate the percentage 
    
2. Add all the above the elements in JSON and write a JSON as 'student_result.json' file
3. Define 'Class' to store all student information and perform the above calculations
4. Add exception handlers (Try... Except) whereever required
5. Use this notebook to type all your codes    

```python
import json


# Store Student objects in the list
studentsList = []
passMarks = 45

# Calculate total mark
def calTotal(marks):
    try:
        return sum(marks)
    except Exception as e:
        print(f'Exception raised in the calTotal function : {e}')
        return 0

# Calculate result based on the pass mark
def calResult(marks):
    try:
        for mark in marks:
            if mark < passMarks:
                result = 'Fail'
                break;
            else:
                result = 'Pass' 
        return result
    except Exception as e:
        print(f'Exception raised in the calResult function : {e}')
        return 'Fail'
    
# Calculate grade
def calGrade(marks):
    try:
        avg = calAvg(marks)
        if avg >= 91:
            grade = 'A+'
        elif avg >= 81 and avg <= 90:            
            grade = 'A'
        elif avg >= 71 and avg <= 80:
            grade = 'B+'
        elif avg >= 65 and avg <= 70:
            grade = 'B'
        elif avg >= 55 and avg <= 64:
            grade = 'C'
        else:
            grade = 'D'
        return grade
    except Exception as e:
        print(f'Exception raised in the calGrade function : {e}')
        return 'D'

# Calculate average
def calAvg(marks):
    try:
        return round(calTotal(marks) / len(marks), 2)
    except Exception as e:
        print(f'Exception raised in the calAvg function : {e}')
        return 0

# Create class "Students"
class Students:   
    # Constructor
    def __init__(self, name=None, gender=None, marks=[]):
        self.name = name
        self.gender = gender
        self.marks = marks
        self.total = None
        self.result = None
        self.grade = None
        self.percentage = None
    
    # Add a student detail as object in the list
    def addStudent(self, student):
        studentsList.append(Students(student['name'], student['gender'], student['marks']))
    
    # Display Student object
    def displayDetail(self):
        print("Name : ", self.name)
        print("Gender : ", self.gender)
        print("Marks : ", self.marks)
        print("Total Mark : ", self.total)
        print("Result : ", self.result)
        print("Grade : ", self.grade)
        print("Percentage : ", self.percentage)
        print("\n")
    
    # Calculate student result
    def calStudentResult(self):
        try:
            self.total = calTotal(self.marks)
            self.result = calResult(self.marks)    
            if self.result == 'Pass':
                self.grade = calGrade(self.marks)
                self.percentage = calAvg(self.marks)
            else:
                self.grade = None
                self.percentage = None
        except Exception as e:
            print(f'Exception raised in the calStudentResult function : {e}')
try:
    # Read a input JSON file
    fileName = 'students.json'
    with open(fileName, 'r') as jsonFile:
        studentData = json.load(jsonFile)

    # Create a empty Students object
    studentObj = Students()    
    # Append all student details in the list
    for student in studentData:
        studentObj.addStudent(student)

    for studObj in studentsList:
        studObj.calStudentResult()
        studObj.displayDetail()
except Exception as e:
    print(f'Exception raised in the main function : {e}')
```
