```python
greenDF = df[(df['alert'] == 'green') & (df['mag'] >= 6.0)]

greenDF
```

```python
with sqlite3.connect('../data/quakes.db') as connection:
    greenDF.to_sql('greentsunamis', connection, index=False, if_exists='replace')
```

```python
with sqlite3.connect('../data/quakes.db') as connection:
    tsunamis = pd.read_sql('SELECT * FROM greentsunamis', connection)
```
