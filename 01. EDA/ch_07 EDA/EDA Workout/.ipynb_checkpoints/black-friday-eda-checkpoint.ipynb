> check for outliers

```python
# z-score
outliers = []

def detect_outliers(data):
    threshold = 3 # 3rd standard deviation
    mean = np.mean(data)
    std = np.std(data)
    
    for i in data:
        z_score = (i-mean)/std
        if np.abs(z_score)>threshold:
            outliers.append(i)
    return outliers

detect_outliers(df['Purchase'])
```

> map range variables into integers (e.g 'Age' column)

```python
df.Age.unique()
```

```python
def rename_age(age):
    if '0-17' in age:
        age = age.replace('0-17','child')
        return str(age)
    elif '18-25' in age:
        age = age.replace('18-25','teenage')
        return str(age)
    elif '26-35' in age:
        age = age.replace('26-35','adult')
        return str(age)
    elif '36-45' in age:
        age = age.replace('36-45','adult')
        return str(age)
    elif '46-50' in age:
        age = age.replace('46-50','mid-age')
        return str(age)
    elif '51-55' in age:
        age = age.replace('51-55','mid-age')
        return str(age)
    else:
        age = age.replace('55+','old')
        return str(age)
    
df.Age = df.Age.apply(rename_age)
```

```python
df.Age = df.Age.map({'child':0,'teenage':1,'adult':2,'mid-age':3,'old':4}).astype(int)
```

1. We have replaced the age groups with respective categories
2. Then map these different categories to integers

- 0-17 : child --> 0
- 18-25 : teenage --> 1
- 26-35 : adult --> 2
- 36-45 : adult --> 2
- 46-50 : mid-age --> 3
- 51-55 : mid-age --> 3
- 55+ : old --> 4

> fill nan values

Product2

```python
s=df.groupby('Age')['Product2'].agg(pd.Series.mode)
for key in s.index:
    df.loc[(df['Age']==key) & (df['Product2'].isnull()), 'Product2']=s[key]
```

Product3

```python
s=df.groupby('Age')['Product3'].agg(pd.Series.mode)
for key in s.index:
    df.loc[(df['Age']==key) & (df['Product3'].isnull()), 'Product3']=s[key]
```
